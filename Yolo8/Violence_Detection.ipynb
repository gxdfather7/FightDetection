```python
# Install ffmpeg if needed
!apt-get install ffmpeg

# Convert .avi to .mp4
input_path = '/content/drive/MyDrive/thesis/dataset/Violence.v1i.yolov11/runs/detect/predict2/newfi96.avi'
output_path = '/content/drive/MyDrive/thesis/dataset/Violence.v1i.yolov11/runs/detect/predict2/newfi96.mp4'

!ffmpeg -i "{input_path}" -vcodec libx264 "{output_path}"

```

```python
from IPython.display import HTML
from base64 import b64encode

# Path to the converted .mp4 video
video_path = output_path

# Read and encode the video file to embed in HTML
with open(video_path, "rb") as video_file:
    video_data = b64encode(video_file.read()).decode()

# Display the video in HTML
HTML(f"""
<video width=600 controls>
      <source src="data:video/mp4;base64,{video_data}" type="video/mp4">
</video>
""")
```
